```python
# Buy Orders
from time import sleep
import krakenex
from pykrakenapi import KrakenAPI
api = krakenex.API()
kraken = KrakenAPI(api)

api.load_key('KrakenPass.txt')

while True:
    
    try:
        BTC_old = float((kraken.get_ticker_information('BTCCAD'))['b'][0][0])
    except exception as e:
        print(f'Unable to obtain BTC data: {e}')
    
    sleep(60)
    
    try:
        BTC_new = float((kraken.get_ticker_information('BTCCAD'))['b'][0][0])
    except exception as e:
        print(f'Unable to obtain BTC data: {e}')
    
    percent = ((BTC_new - BTC_old)*100) / BTC_old
    
    if percent >= 0.0015:
        try:
            BTC = float((kraken.get_ticker_information('BTCCAD'))['a'][0][0]) + 2
            
            response = kraken.add_standard_order(pair='XXBTZCAD', type='buy', ordertype='limit', 
                                                 volume='0.0001', price=BTC, validate=False)
            print(response)
        except exception as e:
            print(f'Error placing order: {e}')
    
        sleep(3)
    
        check_order = kraken.query_orders_info(response['txid'][0])
    
        if check_order['status'][0] == 'open' or 'closed':
            print('Order completed sucessfully')
            break
        else:
            print('Order rejected')
            break
    else:
        print('Requirement not reached')
```
